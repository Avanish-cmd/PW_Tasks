# Q3. Why is redirect() used in Flask?

### The redirect() function is used in Flask to redirect the user to a different URL.
### After a user has successfully logged in to your website, you might want to redirect them to a page where they can access their account information.
### The redirect() function takes a single argument, which is the URL that the user should be redirected to. For example, redirect('/login') would redirect the user to the /login URL.

```python
from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the home page!'

@app.route('/about')
def about():
    return 'This is the about page.'

@app.route('/redirect')
def do_redirect():
    # Redirect the user to the about page
    return redirect(url_for('about'))

if __name__ == '__main__':
    app.run()
```

# Q4. What are templates in Flask? Why is the render_template() function used?

```python
### In Flask, templates are used to generate dynamic HTML pages that can display data from the application. A template is a file that contains a mixture of static HTML and dynamic content that is generated by the application.
### To render a template in Flask, you use the render_template() function. This function takes at least one argument, which is the name of the template file to render. The template file should be located in a subdirectory called templates within the Flask application directory.
```

```python
from flask import Flask, render_template

app = Flask(__name__)

@app.route('/')
def home():
    # Render the home.html template
    return render_template('home.html')

if __name__ == '__main__':
    app.run()

    
# In this example, the home() function renders the home.html template using render_template(). 
# The template file is located in the templates subdirectory of the Flask application directory.
```

# Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

```python
# Flask API that responds to a GET request with a JSON object:
from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/hello')
def hello():
    return jsonify({'message': 'Hello, world!'})

if __name__ == '__main__':
    app.run()
```
